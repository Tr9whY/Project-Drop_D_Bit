```python
import pandas as pd
import matplotlib.pyplot as plt

Pet_병원 = pd.read_csv('동별통합데이터_csv파일/동물병원_Gu+Dong_Count__v1.csv')
Pet_판매 = pd.read_csv('동별통합데이터_csv파일/판매업_Gu_Dong_Counts__v1.csv')
Pet_위탁 = pd.read_csv('동별통합데이터_csv파일/위탁업_Gu_Dong_Counts__v1.csv')
Pet_운송 = pd.read_csv('동별통합데이터_csv파일/운송업_Gu_Dong_Counts__v1.csv')
Pet_생산 = pd.read_csv('동별통합데이터_csv파일/생산업_Gu_Dong_Counts__v1.csv')
Pet_수입 = pd.read_csv('동별통합데이터_csv파일/수입업_Gu_Dong_Counts__v1.csv')
```

```python
def Change_Col_Names(*Variable_Name):
    for k in Variable_Name:
        k.rename(columns={
                        'Count': 'Counts',
                        'count': 'Counts',
                        '시도 명칭':'City',
                                            '시군구 명칭':'Gu',
                                            '법정읍면동명칭' : 'Dong',
                                            '위도' : 'Y',
                                            '경도' : 'X'}, inplace=True)
        k.sort_values(by=['Gu','Dong'], inplace=True)
        k.reset_index(drop = True, inplace=True)

Change_Col_Names(Pet_병원, Pet_판매, Pet_위탁, Pet_운송, Pet_생산, Pet_수입)
```

### 카테고리 세분화입니다.

```python
Pet_병원_col = Pet_병원.rename(columns={'Category':'Hospital','Counts':'Hospital_Counts'})
Pet_판매_col = Pet_판매.rename(columns={'Category':'Sales','Counts':'Sales_Counts'})
Pet_위탁_col  = Pet_위탁.rename(columns={'Category':'Consignment','Counts':'Consignment_Counts'})
Pet_운송_col  = Pet_운송.rename(columns={'Category':'Transit','Counts':'Transit_Counts'})
Pet_생산_col  = Pet_생산.rename(columns={'Category':'Production','Counts':'Production_Counts'})
Pet_수입_col  = Pet_수입.rename(columns={'Category':'Import','Counts':'Import_Counts'})
```

### 카테고리별 컬럼 만들기입니다.

```python
# 컬럼 통일
Pet_병원_col = Pet_병원_col[['Gu','Dong','Hospital_Counts']]
Pet_판매_col = Pet_판매_col[['Gu','Dong','Sales_Counts']]
Pet_위탁_col = Pet_위탁_col[['Gu','Dong','Consignment_Counts']]
Pet_운송_col = Pet_운송_col[['Gu','Dong','Transit_Counts']]
Pet_생산_col = Pet_생산_col[['Gu','Dong','Production_Counts']]
Pet_수입_col = Pet_수입_col[['Gu','Dong','Import_Counts']]
```
